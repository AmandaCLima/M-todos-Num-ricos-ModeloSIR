#Regimes de Vacina√ß√£o üè•  
Ser√£o descritos quatro tipos diferentes de esquemas de vacina√ß√£o, sendo dois deles est√°ticos, ou seja, a taxa de vacinas disponibilizadas n√£o √© alterada ao longo de um ano, e dois din√¢micos, quando essa taxa sofre varia√ß√µes nesse mesmo per√≠odo.



> **Regime Est√°tico**



*   Constante
  
  A principal caracter√≠stica desse regime, √© a distribui√ß√£o equitativa das vacina√ß√µes, ou seja, a mesma propor√ß√£o de indiv√≠duos √© imunizada em cada cidade, independentemente de sua classifica√ß√£o ou densidade populacional.
  Por√©m, ainda que essa estrat√©gia possua uma aplica√ß√£o simples e direta, ela n√£o √© t√£o pr√°tica quando o acesso aos locais mais distantes das cidades grandes √© levado em considera√ß√£o.


*   Por Cidade

  Diferentemente do regime est√°tico constante, o regime est√°tico por cidade consiste em uma estrat√©gia de vacina√ß√£o na qual uma propor√ß√£o diferente de indiv√≠duos √© imunizada a depender da classe de uma cidade, de modo que os locais com maior densidade demogr√°fica recebam as maiores taxas de vacina√ß√£o.

  A l√≥gica por tr√°s desse esquema consiste no fato de que a dissemina√ß√£o de doen√ßas ocorre mais rapidamente em popula√ß√µes mais densas do que em cidades pequenas e m√©dias. Al√©m disso, o maior acesso aos cuidados de sa√∫de que se d√° nos grandes centros garante maior praticidade a esse regime.



> **Regime Din√¢mico**



*   Por Tempo

  Nesse regime, a varia√ß√£o da taxa de vacina√ß√£o ocorre de acordo com as flutua√ß√µes naturais dos riscos de infec√ß√£o ao longo do ano, de modo que sejam aumentadas as campanhas de imuniza√ß√µes meses antes do pico de uma determinada doen√ßa, evitando a propaga√ß√£o da mesma. Logo, ao acompanhar essas varia√ß√µes sazonais, esse esquema permite uma resposta eficaz e direcionada √† amea√ßa de certas enfermidades ao longo do ano.


*   Por Infectados

  A caracter√≠stica que marca esse regime √© o fato de a taxa de vacina√ß√µes acompanhar a quantidade de pessoas infectadas em uma regi√£o, dessa forma, quanto maior a dissemina√ß√£o de uma doen√ßa em uma determinada cidade, mais intensas ser√£o as campanhas de imuniza√ß√£o nela. Isso busca n√£o s√≥ evitar que os casos cres√ßam ainda mais, como tamb√©m agir como um lembrete da import√¢ncia da vacina√ß√£o na preven√ß√£o e combate de doen√ßas.

  No entanto, √© importante ressaltar que essa abordagem reativa muitas vezes n√£o √© t√£o eficaz quanto programas de vacina√ß√£o cont√≠nuos e abrangentes instaurado previamente a surtos de doen√ßas.




# Modelo SIR ü¶†
Para realizar as previs√µes, ser√° utilizado uma varia√ß√£o do modelo epidemiol√≥gico SIR (Suscet√≠vel-Infectado-Recuperado) que inclui a vacina√ß√£o. Esse modelo, apesar de simples √© considerado robusto. No entanto, o que o diferencia √© o fato dele considerar que um indiv√≠duo possa se recuperar e ficar imune a uma determinada doen√ßa, sendo, portanto, um modelo amplamente utilizado para descrever a evolu√ß√£o de doen√ßas epidemiol√≥gicas.Consideramos tr√™s eventos poss√≠veis:  a infec√ß√£o, a vacina√ß√£o e a recupera√ß√£o da doen√ßa. Al√©m disso, consideramos cada unidade de tempo equivalente a 2 semanas e o tempo de recupera√ß√£o tamb√©m igual a duas semanas.

* S(t) - pessoas suscet√≠veis √† infec√ß√£o
* I(t) - pessoas infectadas pela doen√ßa
* R(t) - pessoas recuperadas ou que se tornaram imunes atrav√©s da vacina√ß√£o
* N(t) = S(t) + I(t) + R(t) : Popula√ß√£o total
* Œ≤: Percentual Indiv√≠duos infectados em 2 semanas
* œÜ: N√∫mero de indiv√≠duos vacinados suscet√≠veis em 2 semanas

* œâ: Taxa de recupera√ß√£o = \begin{equation}
    \frac{1}{T de Recupera√ß√£o} = 1
\end{equation}



Assim,o sistema de equa√ß√µes diferenciais ser√° este:

\begin{equation}
    \frac{dS}{dt} = -Œ≤ S  I  - œÜ S
\end{equation}

\begin{equation}
    \frac{dI}{dt} = Œ≤  S  I - œâ I
\end{equation}

\begin{equation}
    \frac{dR}{dt} = œÜ S + œâ I
\end{equation}

 * O Œ≤ ir√° variar dependendo se a cidade escolhida √© uma capital, uma cidade ou um centro regional, al√©m de variar em rela√ß√£o ao tempo. Al√©m disso, ele vai variar com o par√¢metro $Œ∏$, que est√° relacionado com a √©poca do ano. Foi escolhido o per√≠odo de f√©rias escolares, onde as intera√ß√µes entre as crian√ßas s√£o menores, resultando em $Œ∏$ = 16,5.Tal informa√ß√£o foi retirada do artigo de Bj√∏rnstad sobre a estimativa de tais taxas relacionadas ao sarampo(Bj√∏rnstad, 2002).Essa √© a f√≥rmula que ir√° determinar o Œ≤:
 \begin{equation}
    Œ≤ = \overline{Œ≤} - 0,4 \overline{Œ≤} sin(\frac{(t+œ¥)œÄ}{13})
\end{equation}

* O œÜ ir√° variar dependendo do regime escolhido: Est√°tico constante, est√°tico por cidade, din√¢mico por tempo ou din√¢mico por infectados.Os valores foram determinados de acordo com o artigo de Emily Nguyen.
* Se for est√°tico constante, os valores de œÜ ser√£o: 0.06, 0.08, 0.1, 0.12, 0.14 e 0.16.
* Se for est√°tico por cidade iremos analisar tr√™s regimes diferentes. O primeiro h√° uma vacina√ß√£o somente das capitais, o segundo uma distribui√ß√£o n√£o igualit√°ria, dando pref√™ncia para as capitais e em siguida os centros regionais e por √∫ltimo uma distribui√ß√£o mais realista das vacina√ß√µes.

#Visualizando os Dados üëÅ


```python
#importando os pacotes necess√°rios para modelagem da ODE

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import plotly.express as px
```

```python
#declarando as vari√°veis que s√£o usadas no modelo
N = 10000

#n√∫mero inicial de pessoas suscet√≠veis, infectadas e recuperadas
S0 = N -1
I0 = 1
R0 = 0

#taxa de propaga√ß√£o da doen√ßa (para Cidades)
B = 0.0031

#taxa de imuniza√ß√£o (Utilizando a taxa de imuniza√ß√£o constante (Regime 3))
phi = 0.03

#taxa de recupera√ß√£o
w = 1

valores = S0, I0, R0, B, phi, w
t = np.linspace(0, 14, 14)
```

```python
# c√°lculo do modelo

def deriv (y, t, N, valores):
  S, I, R = y
  beta = B *(1 - 0.4* np.sin(np.pi*(t+16.5)/13))
  dSdt = -(beta*S*I) - (S*phi)
  dIdt = (beta*S*I) - (w*I)
  dRdt = (phi * S) + (w*I)
  return dSdt, dIdt, dRdt

#condi√ß√µes iniciais
y0 = valores[0], valores[1], valores[2]
ret = odeint(deriv, y0, t, args = (N, valores))
S, I, R = ret.T
```

```python
# essa fun√ß√£o plota o gr√°fico

def plotMyGraph(S, I, R, t):
  fig = plt.figure(facecolor='w')
  ax = fig.add_subplot(111, axisbelow=True)

  ax.plot(t, S, 'b', alpha=0.7, lw=2, label='S')
  ax.plot(t, I, 'r', alpha=0.7, lw=2, label='I')
  ax.plot(t, R, 'y', alpha=0.7, lw=2, label='R')

  ax.set_xlabel('2 semanas')
  ax.set_ylabel('Popula√ß√£o')
  plt.title("Evolu√ß√£o do Sarampo em cidades com taxa de imuniza√ß√£o constante (Regime 3)")
  legend = ax.legend(loc='best')
  legend.get_frame().set_alpha(0.5)

  plt.show()
```

```python
plotMyGraph(S, I, R, t)
```
